## Pre-processing

### Grouping fitur yang mirip

```python
#terlihat ada beberapa fitur yang sama, tetapi terpisah (Age, Gender, Severity, Contact)

#terlihat ada beberapa fitur yang sama, tetapi terpisah (Age, Gender, Severity, Contact)

severity_columns = data.filter(like='Severity_').columns

data['Severity_None'].replace({1:'None',0:'No'},inplace =True)
data['Severity_Mild'].replace({1:'Mild',0:'No'},inplace =True)
data['Severity_Moderate'].replace({1:'Moderate',0:'No'},inplace =True)
data['Severity_Severe'].replace({1:'Severe',0:'No'},inplace =True)

data['Condition'] = data[severity_columns].values.tolist()

def removing(list1):
    list1 = set(list1) 
    list1.discard("No")
    a = ''.join(list1)
    return a

data['Condition'] = data['Condition'].apply(removing)


age_columns = data.filter(like='Age_').columns

data['Age_0-9'].replace({1:'0-9',0:'No'},inplace =True)
data['Age_10-19'].replace({1:'10-19',0:'No'},inplace =True)
data['Age_20-24'].replace({1:'20-24',0:'No'},inplace =True)
data['Age_25-59'].replace({1:'25-59',0:'No'},inplace =True)
data['Age_60+'].replace({1:'60+',0:'No'},inplace =True)

data['Age'] = data[age_columns].values.tolist()

def removing(list2):
    list2 = set(list2) 
    list2.discard("No")
    b = ''.join(list2)
    return b

data['Age'] = data['Age'].apply(removing)

gender_columns = data.filter(like='Gender_').columns

data['Gender_Female'].replace({1:'Female',0:'No'},inplace =True)
data['Gender_Male'].replace({1:'Male',0:'No'},inplace =True)
data['Gender_Transgender'].replace({1:'Transgender',0:'No'},inplace =True)

data['Gender'] = data[gender_columns].values.tolist()

def removing(list3):
    list3 = set(list3) 
    list3.discard("No")
    c = ''.join(list3)
    return c

data['Gender'] = data['Gender'].apply(removing)


contact_columns = data.filter(like='Contact_').columns

data['Contact_Dont-Know'].replace({1:'Dont-Know',0:'No'},inplace =True)
data['Contact_No'].replace({1:'Not',0:'No'},inplace =True)
data['Contact_Yes'].replace({1:'Yes',0:'No'},inplace =True)

data['Contact'] = data[severity_columns].values.tolist()

def removing(list4):
    list4 = set(list4) 
    list4.discard("No")
    d = ''.join(list4)
    return d

data['Contact'] = data['Contact'].apply(removing)

data
```

```python
import keras
import tensorflow as tf

!pip install -q -U keras-tuner

import keras_tuner as kt
```

```python
def build_model(hp):
  model = keras.models.Sequential() #kita inisialisasi terlebih dahul model sequential
  

  hp_layers = hp.Choice('layer', [3,10]) # kita mulai mendefinisikan variabel
  hp_neurons = hp.Int('neuron', min_value = 100, max_value = 500, step = 50)
  #terdapat 2 variabel yaitu hp layer dan neurons. hp.choice maksudnya ialah ingin memvariasikan nilai layer nya itu bisa 2 atau 3
  #hp.int itu akan memvariasikan nilai neuron dengan minimal value nya 100 dan max nya 300 step nya 200. Artinya nilai2 yang mungkin pada hp neuron itu 100 200 dan 300.
  #disini kita akan memiliki beberapa kombinasi, (cont : kita punya arsitekturnya itu 2 hidden layer dan disetiap hidden layer nya itu punya 100 neuron atau bisa jg kita punya 3 hiden layer dengan 200 neuron)

  for layers in range(hp_layers): #kalo misalkan hp layersnya itu 3 maka dia akan mengulang 3 kali
    model.add(keras.layers.Dense(hp_neurons, activation = 'relu')) #kita akan mebuat 3 hidden layer(model.add) dimana banyaknya neuronnya itu (hp neuron)
  model.add(keras.layers.Dense(10, activation = 'softmax')) #output layer nya ada 10 neuron dan 20 nilai output

  #arsitektur model nya sudah selesai dibuat lalu kita akan lakukan kompilasi

  optimizer = tf.keras.optimizers.Adam()
  model.compile(loss = 'sparse_categorical_crossentropy',
                 optimizer = optimizer,
                 metrics = ['accuracy'])
  return model
```

```python
##sekarang akan dibuat hyperparameter turning nya
#hyperband tuner
tuner = kt.Hyperband(build_model,
                     objective = 'val_accuracy',
                     max_epochs = 25,
                     directory = 'my_di',
                     project_name = 'my_tuner') #disimpan dalam var tuner
#penjelasannya = objektif nya menggunakan akurasi pada data validasi (yang akan menjadi tolak ukur dari keakuratan
#tuning), max epochs nya, directory (lokasi dmn file tuning akan disimpan (namainnya terserah apa aja))
```

```python
stop_early = tf.keras.callbacks.EarlyStopping(
    monitor = 'val_loss',
    patience = 5
)
```

```python
#proses pencarian hyperparameter
tuner.search(X_trainX, y_trainX,
             epochs = 25,
             validation_data = [X_valid, y_valid],
             callbacks = [stop_early])
```

```python
best_hps = tuner.get_best_hyperparameters()[0]

print("Best number of layers: ", best_hps.get("layer"))
print("Best number of neuron: ", best_hps.get("neuron"))
print("Best number of epochs: ", best_hps.get("tuner/epochs"))
```

```python
model = tuner.hypermodel.build(best_hps) #build_model 
```

```python
model.fit(X_trainX, y_trainX,
          epochs = 3,
          validation_data = [X_valid, y_valid])
```

```python
#memprediksi dengan model
model.evaluate(X_test, Y_test)
```
