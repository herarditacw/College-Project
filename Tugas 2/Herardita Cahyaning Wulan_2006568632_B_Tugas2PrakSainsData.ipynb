## Pre-processing data

```python
#memilih fitur yang sekiranya berpengaruh pada kemungkinan heart disease. saya pilih 12 fitur dan 1 target

heart_new = heart_df[['age', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh',
       'exng', 'oldpeak', 'slp', 'caa', 'thall', 'output']]
heart_new.info()
```

## Train-test-validasi Data

```python
#menyimpan fitur2 berpengaruh pada array X dan target pada array Y
X = heart_new[['age', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh',
       'exng', 'oldpeak', 'slp', 'caa', 'thall']].values
Y = heart_new['output']

 #menampilkan X dan Y
print('X :\n', X)
print('\nY :\n', Y)
```

```python
#membagi data menjadi train, test, valid data
from sklearn.model_selection import train_test_split #import package

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 3) #train : test = 8 : 2. pilih random_state = 3 agar data train dan test tidak berubah saat run kembali

#ambil data validasi sebanyak 180 < 242 (data train) dan pengambilan data 30 30
X_valid, X_trainX = X_train[:180]/30.0, X_train[180:]/30.0
y_valid, y_trainX = Y_train[:180], Y_train[180:]

#menampilkan train, test, validasi data
print('Train Set : ', X_train.shape, Y_train.shape)
print('Test Set : ', X_test.shape, Y_test.shape)
print('Validation Set : ', X_valid.shape, y_valid.shape)
```

## Membuat Model

```python
pip install keras #install keras
```

```python
#import package yang akan digunakan

import keras
import tensorflow as tf

!pip install -q -U keras-tuner

import keras_tuner as kt
```

```python
#mendefinisikan model yang akan dibuat

def build_model(hp):
  model = keras.models.Sequential() #inisialisasi model Sequential

  hp_layers = hp.Choice('layer', [2,3]) #pemilihan layer terbaiknya 2 atau 3, karena pd soal diminta DNN jadi layer harus > 1
  hp_neurons = hp.Int('neuron', min_value = 100, max_value = 500, step = 10) #pemilihan neuron terbaik dengan range [100,500] dan stepsizenya 10
  
  for layers in range(hp_layers): #iterasi layer sesuai dengan hasil pemilihan layer terbaik
    model.add(keras.layers.Dense(hp_neurons, activation = 'tanh')) #menambahkan hidden layer sesuai jumlah layer. menggunakan fungsi aktivasi tanh karena setelah dibandingkan dgn relu, model dgn fungsi tanh lebih baik hasilnya
  model.add(keras.layers.Dense(10, activation = 'softmax')) #menampilkan outputnya

  optimizer = tf.keras.optimizers.Adam() #fungsi optimisasi menggunakan Adam karena menurut beberapa jurnal, fungsi Adam memiliki berbagai kelebihan
  model.compile(loss = 'sparse_categorical_crossentropy',
                 optimizer = optimizer,
                 metrics = ['accuracy'])
  return model #memanggil model yg sudah didefinisikan tadi
```

## Hyperparameter Tuning

```python
#untuk melakukan hyperparameter tuning, bentuk hyperband tuner nya dulu
tuner = kt.Hyperband(build_model,
                     objective = 'val_accuracy', #tolak ukur dari keakuratan tuningnya
                     max_epochs = 500, #maksimal epochs di setiap layer
                     directory = 'my_di', #lokasi penyimpanan file tuning
                     project_name = 'my_tuner') #lokasi penyimpanan file tuning
```

```python
#untuk memberhentikan trial jika proses trial dirasa tidak akan mendapat hasil bagus

stop_early = tf.keras.callbacks.EarlyStopping( 
    monitor = 'val_loss',
    patience = 5)
```

```python
#mencari hyperparameter terbaik
tuner.search(X_trainX, y_trainX, #data yg digunakan untuk melatih model dlm hyperparameter tuning
             epochs = 500, #maksimal epochs
             validation_data = [X_valid, y_valid], #data yg digunakan untuk memvalidasi data train
             callbacks = [stop_early]) #untuk memberhentikan proses trial jika dirasa trial tidak mendapat hasil signifikan
```

```python
best_hps = tuner.get_best_hyperparameters()[0] #menampilkan hyperparameter terbaik utnuk model ini

print("Best number of layers: ", best_hps.get("layer"))
print("Best number of neuron: ", best_hps.get("neuron"))
print("Best number of epochs: ", best_hps.get("tuner/epochs"))
```

```python
model = tuner.hypermodel.build(best_hps) #membentuk model dengan hyperparameter terbaik
```

```python
#membuat model dengan menyesuaikan dgn data train dan data validasi

model.fit(X_trainX, y_trainX,
          epochs = 3,
          validation_data = [X_valid, y_valid])
```

## Evaluasi Model

```python
y_proba = model.predict(X_test) #melihat probabilitas model

y_p = np.array([np.argmax(i) for i in y_proba]) #memprediksi model dengan data test
#print(y_p) untuk menampilkan hasil prediksi
```

```python
#menghitung nilai akurasi, recall, presisi, dan F1 scorenya

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score #import package yang akan digunakan

#menampilkan setiap nilainya
print('Accuracy :',(accuracy_score(Y_test, y_p)))
print('Recall:',(recall_score(Y_test, y_p)))
print('Precision:',(precision_score(Y_test, y_p)))
print('F1-Score:',(f1_score(Y_test, y_p)))
```

**Analisis Model**

Model yang sudah dibuat adalah *Deep Neural Network* dengan 3 layer, 210 neuron, dan 3 epochs tiap layernya. Parameter tsb. dihasilkan dari proses hyperparameter tuning.


Setelah model terbentuk, dilakukan evaluasi model yaitu memprediksi Y_test dengan model yg sudah dibuat. Langkah terakhir adalah menampilkan nilai akurasi, recall, presisi, dan F1 score. Dari 4 nilai tersebut, dapat dilihat bahwa semua nilainya bernilai cukup besar, bahkan nilai recall mencapai 1. Artinya, model yg dibuat terbilang bagus dan cukup akurat.


Model ini digunakan untuk menyelesaikan masalah klasifikasi risiko penyakit jantung. Oleh karena itu, lebih baik kita menggunakan model yg memiliki nilai recall tinggi, karena nilai recall bergantung pada poin False Negative, artinya model memprediksi pasien terkena penyakit jantung padahal tidak terkena penyakit jantung. Hal ini lebih baik daripada model mendeteksi seseorang tidak terkena penyakit jantung padahal terkena penyakit jantung.

## Membuat Model

```python
pip install keras #install keras
```

```python
#import package yang akan digunakan untuk membuat model

import keras
import tensorflow as tf

!pip install -q -U keras-tuner

import keras_tuner as kt
```

```python
#mendefinisikan model yang akan dibuat

def build_model(hp):
  model = keras.models.Sequential() #inisialisasi model Sequential

  hp_layers = hp.Choice('layer', [3,4]) #pemilihan layer terbaiknya 3 atau 4, karena pd soal diminta DNN jadi layer harus > 1
  hp_neurons = hp.Int('neuron', min_value = 100, max_value = 500, step = 10) #pemilihan neuron terbaik dengan range [100,500] dan stepsizenya 10
  
  for layers in range(hp_layers): #iterasi layer sesuai dengan hasil pemilihan layer terbaik
    model.add(keras.layers.Dense(hp_neurons, activation = 'tanh')) #menambahkan hidden layer sesuai jumlah layer. menggunakan fungsi aktivasi tanh karena setelah dibandingkan dgn relu, model dgn fungsi tanh lebih baik hasilnya
  model.add(keras.layers.Dense(10, activation = 'softmax')) #menghasilkan output


  optimizer = tf.keras.optimizers.Adam() #fungsi optimisasi menggunakan Adam karena menurut beberapa jurnal, fungsi Adam memiliki berbagai kelebihan
  model.compile(loss = 'sparse_categorical_crossentropy',
                 optimizer = optimizer,
                 metrics = ['accuracy']) #penilaian model dengan menampilkan akurasinya
  return model #memanggil model yg sudah didefinisikan tadi
```

## Hyperparameter tuning

```python
#bentuk hyperband sebelum melakukan hyperparameter tuning
tuner = kt.Hyperband(build_model,
                     objective = 'val_accuracy',#tolak ukur dari keakuratan tuningnya
                     max_epochs = 500, #maksimal epochs di setiap layer
                     directory = 'my_di', #lokasi penyimpanan file tuning
                     project_name = 'my_tuner') #lokasi penyimpanan file tuning
```

```python
#untuk memberhentikan trial jika proses trial dirasa tidak akan mendapat hasil bagus

stop_early = tf.keras.callbacks.EarlyStopping(
    monitor = 'val_loss',
    patience = 5)
```

```python
#proses pencarian hyperparameter terbaik
tuner.search(X_trainX, y_trainX, #data yg digunakan untuk melatih model dlm hyperparameter tuning
             epochs = 500, #maksimal epochs
             validation_data = [X_valid, y_valid], #data yg digunakan untuk memvalidasi data train
             callbacks = [stop_early]) #untuk memberhentikan proses trial jika dirasa trial tidak mendapat hasil signifikan
```

```python
#menampilkan hyperparameter terbaik untuuk model ini

best_hps = tuner.get_best_hyperparameters()[0]

print("Best number of layers: ", best_hps.get("layer"))
print("Best number of neuron: ", best_hps.get("neuron"))
print("Best number of epochs: ", best_hps.get("tuner/epochs"))
```

```python
#membuat model dengan menyesuaikan dgn data train dan data validasi

model.fit(X_trainX, y_trainX,
          epochs = 3,
          validation_data = [X_valid, y_valid])
```

```python
#karena dari hasil korelasi terlihat bahwa veil-type menghasilkan NaN, maka cek data pd fitur veil-type. ternyata fitur ini hanya berisi satu jenis data
#maka langsung hilangkan secara manual dari list fiturkarena tidak berpengaruh
jamur_df['veil-type'].unique()
```

```python
#menghitung korelasi tiap fitur yg baru

corr_matrix = jamur_fitur.corr().abs()
```

```python
#mengambil matriks segitiga atas

upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
```

```python
#menghapus fitur yang kurang berkorelasi dengan fitur lain

to_drop = [column for column in upper.columns if any(upper[column] > 0.6)]
for ls in to_drop :
  jamur_fitur = jamur_fitur.drop([ls], axis = 1)
```
